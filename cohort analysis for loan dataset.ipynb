```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
loans = pd.read_csv("lending_club_loan_two.csv", low_memory =  False)
loans
```

```python
loans.columns
```

```python
loans['issue_d']
```

```python
loans['issue_d'][0]
```

```python


# Convert the 'date_string' column to datetime objects
loans['x'] = pd.to_datetime(loans['issue_d'], format='%b-%Y')

loans['x']
```

```python
loans['month'] = loans['x'].dt.month
```

```python
loans['month']
```

```python
loans['issue_d'] = loans['issue_d'].apply(lambda x: datetime.strptime(x, '%b-%Y').year)
```

```python
loans['issue_d'] = loans['issue_d'].apply(lambda x: datetime.strptime(str(x), '%Y'))
```

```python
loans['issue_d'].unique()
```

```python
loans['issue_d']
```

```python
loans['Year'] = loans['issue_d'].dt.to_period("Y")
```

```python
loans['Year']
```

```python
loans.columns
```

```python
loans['installment'] = loans['installment'] * 12
loans['installment']
```

```python
cohort_data = loans.groupby(['emp_title', 'month'])["installment"].sum().reset_index()
cohort_data
```

```python
# Create a pivot table to get the loan amount sum by cohort month and loan origination month
cohort_data = cohort_data.pivot(index="emp_title", columns="month", values="installment")
cohort_data
```

```python
cohort_data["cohort_month"] = cohort_data.iloc[:,0].name
```

```python
cohort_data["cohort_month"]
```

```python
cohort_data["cohort_index"] = np.arange(1, len(cohort_data)+1)
cohort_data["cohort_index"]
```

```python
cohort_size = cohort_data.groupby("cohort_month").size()
repayment = cohort_data.groupby(["cohort_month", "cohort_index"]).size()
repayment = repayment.divide(cohort_size, axis=0)
```
